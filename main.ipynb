```python
import tracemalloc

def run_experiment():
    # אתחול המדידה
    tracemalloc.start()
    
    print("--- ניסוי 1: שימוש ב-List Comprehension ---")
    # יצירת רשימה שלמה בזיכרון
    squares_list = [i**2 for i in range(10_000_000)]
    result_1 = sum(squares_list)
    
    current, peak = tracemalloc.get_traced_memory()
    print(f"זיכרון בשיא: {peak / 10**6:.2f} MB")
    tracemalloc.stop()

    print("\n--- ניסוי 2: שימוש ב-Generator Expression ---")
    tracemalloc.start()
    # שימוש בסוגריים עגולים - לא יוצר רשימה, אלא מחשב 'תוך כדי תנועה'
    squares_gen = (i**2 for i in range(10_000_000))
    result_2 = sum(squares_gen)
    
    current, peak = tracemalloc.get_traced_memory()
    print(f"זיכרון בשיא: {peak / 10**6:.2f} MB")
    tracemalloc.stop()

run_experiment()
```
